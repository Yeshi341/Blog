```python
def early_shift_emp(emps):
    shift_emp = []
    for i in range(len(emps)):
        if emps['avail_time'][i][0] <= 8:
            shift_emp.append(emps['emp_id'][i])
            i +=1
    return shift_emp

def second_shift_emp(emps):
    shift_emp = []
    for i in range(len(emps)):
        if emps['avail_time'][i][0] > 8:
            shift_emp.append(emps['emp_id'][i])
            i +=1
    return shift_emp
early_shifter = early_shift_emp(employees_df)
second_shifter = second_shift_emp(employees_df)
```
